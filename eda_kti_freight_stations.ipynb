```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
path_train = r"./train/"

kti_data = pd.read_parquet(path_train + '/kti_izm.parquet').convert_dtypes()
freight_data = pd.read_parquet(path_train + '/freight_info.parquet').convert_dtypes()
stations_data = pd.read_parquet(path_train + '/stations.parquet').convert_dtypes()
```

#### Распределение толщины гребня/обода

```python
fig, axes = plt.subplots(2, 4, figsize=(20,8))
types = ['flange', 'rim']
for x in range(2):
    for i in range(1, 5):
        col = f'axl{i}_l_w_{types[x]}'
        kti_data[col].plot.hist(ax=axes[x][i - 1], bins=100, title=col)
```

#### Разность между левым и правым колесом

```python
fig, axes = plt.subplots(2, 4, figsize=(20,8))
types = ['flange', 'rim']
for x in range(2):
    for i in range(1, 5):
        (kti_data[f'axl{i}_l_w_{types[x]}'] - kti_data[f'axl{i}_r_w_{types[x]}']).plot.hist(ax=axes[x][i - 1], bins=100, title=types[x] + ' ' + str(i))
```
